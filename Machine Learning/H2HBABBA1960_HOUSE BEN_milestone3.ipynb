REMOVING DUPLICATE COLUMNS

```python
def DuplicateCols(data1960):
    dupcol=set()
    for x in range(data1960.shape[1]):
        col_1=data1960.iloc[:,x]
        for y in range(x+1,data1960.shape[1]):
            col_2=data1960.iloc[:,y]
            if col_1.equals(col_2):
                dupcol.add(data1960.columns.values[x])
    return list(dupcol)
```

REMOVING QUASI CONSTANTS

```python
dropping=DuplicateCols(data1960)
dropping
```

```python
data1960.drop(columns=dropping,inplace=True) #dropping the same columns
```

ADDING THE REQUIRED COLUMNS FOR CLARIFICATION

```python
Pred_Date['Customer Name'] = tst['name_customer']
```

```python
Pred_Date['Customer #'] = tst['cust_number']
```

```python
Pred_Date['Invoice #'] = tst['invoice_id']
```

```python
Pred_Date['Invoice Amount'] = tst['total_open_amount']
```

```python
Pred_Date['Due Date']=tst['due_in_date']
```

```python
#appending delay column in Pred_Date
delay=[]
for x in predicted:
    delay.append(pd.Timedelta(days=x))
Pred_Date['Delay'] = delay
Pred_Date
```

```python
#Predicted date retrieval
Pred_Date['Predicted Payment Date'] = tst['due_in_date']+Pred_Date['Delay']
Pred_Date
```

ADDING THE AGING BUCKET FACILITY.

```python
aging_bucket = []  #applying aging bucket column in the final prediction
for x in predicted:
    if x<=15:
        aging_bucket.append("0-15days")
    elif x>15 and x<=30:
        aging_bucket.append("16-30days")
    elif x>30 and x<=45:
        aging_bucket.append("31-45days")
    elif x>45 and x<=60:
        aging_bucket.append("46-60days")
    else:
        aging_bucket.append("Greater than 60 days")
Pred_Date['Aging Bucket']= aging_bucket
Pred_Date.drop(['Delay'],axis=1,inplace=True)
Pred_Date
```
